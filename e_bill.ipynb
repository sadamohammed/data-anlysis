```python
#import libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
```

```python
#upload csv file
df=pd.read_csv("/content/electricity.csv")
df.head(18)
```

```python
#saving the csv file
data=df.to_csv("/content/electricity.csv")
```

```python
#creating year and amount columns
column=['Year','Amount']
df_req=df[column]
df_req
```

```python
#dropping null values
df1=df_req.dropna()
```

```python
df1
```

```python
#generate sample data
np.random.seed(0)
df1 = pd.DataFrame({
    'Year': np.random.choice(['2020', '2021', '2022'],size=18),
    'Amount':np.random.normal(loc=10, scale=2, size=18)#loc=>mean,scale=>standard deviation
})
print(df1)
```

```python
#claculate mean amount for each year
mean_score_per_group = df1.groupby('Year')['Amount'].mean()
print(mean_score_per_group)
```

```python
 #generate random year selections
choices = ['2020', '2021', '2022']
size = 18
year_column = np.random.choice(choices, size=size)
print(year_column)
```

```python
#calculate amount values
mean = 10
std = 2
size = 18
Amount_values = mean + std * np.random.randn(size)
print(Amount_values)
```

```python
# hypothesis testing (t-test)
group_A = df1[df1['Year'] == '2020']['Amount']
group_B = df1[df1['Year'] == '2021']['Amount']
t_stat, p_value =stats.ttest_ind(group_A, group_B)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
#prediction
if p_value < 0.05:
  print("reject the null hypothesis")
else:
  print("fail to reject the null hypothesis")
```

```python
#ANOVA (one-way ANOVA)
f_stat, p_value = stats.f_oneway(df1[df1['Year']== '2020']['Amount'],
                                 df1[df1['Year']== '2021']['Amount'],
                                 df1[df1['Year']== '2022']['Amount'])
print("F-statistic:",f_stat)
print("p-value:",p_value)
```
